```python
import math
import numpy as np
import matplotlib.pyplot as plt
import random
import pandas as pd
```

```python
# Simulate a capacitor

def charge(time, C, V, R):

    return V * (1 - math.exp(-t / (R * C)))



def discharge(time, C, V, R):

    return V * math.exp(-t / R*C)
```

```python
# Simulating capacitor charge

vc = []
timec = []

for t in np.arange(0, 10, 0.01): # in seconds

    v = charge(t, 1, 1, 1)

    vc.append(v)

    timec.append(t)
```

```python
# Simulating capacitor discharge

vd = []
timed = []

for t in np.arange(0, 10, 0.01): # in seconds

    v = discharge(t, 1, 1, 1)

    vd.append(v)

    timed.append(t)
```

```python
# Showing graphic of the charge and discharge over the time

def plot_trace_graph(v, time, mode):

    x_coords = time
    y_coords = v
    
    plt.plot(x_coords, y_coords, marker='o', linestyle='-', color='b')
    plt.xlabel('Time (S)')
    plt.ylabel('Voltage (V)')
    plt.title('Capacitor ' + mode)
    plt.grid(True)
    plt.show()

plot_trace_graph(vc, timec, 'charge') # Charge

plot_trace_graph(vd, timed, 'discharge') # Discharge
```
